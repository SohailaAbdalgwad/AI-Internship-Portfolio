# **Support Vector Machine (SVM)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
```

```python
cancer
```

```python
print('Features: ', cancer.feature_names)
print('Labels: ', cancer.target_names)
```

```python
cancer.data.shape #30 feature, the target is not counted
```

```python
print(cancer.data[0:5])
```

```python
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear') #from svm choose support vector classifier, kernel is the method of separation (linear, polynomial, gaussian, polynomial, RBF 'circle it' )
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test,y_predict)
print("Accuracy: %f" % accuracy)
precision = metrics.precision_score(y_test,y_predict)
print("Precision: %f" % precision)
```

```python
recall = metrics.recall_score(y_test,y_predict)
print("Recall: %f" % recall) # medical data
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model_reg = LogisticRegression()
model_reg.fit(x_train, y_train)
```

```python
y_predict_reg = model_reg.predict(x_test)
recall_reg = metrics.recall_score(y_test, y_predict_reg)
print("Recall: %f" % recall_reg)
```

# **Decision Tree**

```python
from sklearn import tree
model_tree = tree.DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
```

```python
y_predict_tree = model_tree.predict(x_test)
recall_tree = metrics.recall_score(y_test,y_predict_tree)
print("Recall: %f" % recall_tree)
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
model_forest = RandomForestClassifier(n_estimators=7 , random_state=0, max_depth= 5)
model_forest.fit(x_train, y_train)
```

```python
y_predict_rf = model_forest.predict(x_test)
recall_rf = metrics.recall_score(y_test, y_predict_rf)
print("Recall: %f" % recall_rf)
```

# **XGB**

```python
from xgboost import XGBClassifier
model_xgb = XGBClassifier()
model_xgb.fit(x_train, y_train)
```

```python
y_predict_xgp = model_xgb.predict(x_test)
recall_xgb = metrics.recall_score(y_test, y_predict_xgp)
print("Recall: %f" % recall_xgb)
```
