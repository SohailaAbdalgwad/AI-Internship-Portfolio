```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
data = np.c_[cancer.data, cancer.target]
columns = np.append(cancer.feature_names, ["target"]) #concat the target with the feature columns
df = pd.DataFrame(data, columns=columns)
x = df.iloc[:, 0:30].values
print(df.columns)
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")

kmeans= KMeans(n_clusters=2)
kmeans.fit(x)
```

```python
print(kmeans.cluster_centers_.shape)
```

```python
x_test =x[7]
```

```python
y_predict = kmeans.predict([x_test])
print(y_predict)
```

```python
print(kmeans.inertia_) #how many iterations happend in the centroid
```

```python
all_inertia=[]

for i in range (1,11) :
  kmeans = KMeans (n_clusters = i)
  kmeans.fit(x)
  all_inertia.append(kmeans.inertia_)

# Plot the graph to visualize the Elbow Method to find the optimal number of cluster
plt.plot (range (1,11), all_inertia)
plt. title('The Elbow Method')
plt.xlabel (' Number of clusters')
plt.ylabel("inertia")
plt.show()
```
