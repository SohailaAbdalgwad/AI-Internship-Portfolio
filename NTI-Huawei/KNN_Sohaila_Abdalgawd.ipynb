```python
df = pd.json_normalize(data, meta=['id', 'height', 'age', 'weight'])
df
```

```python
df_train = df
df_train = df_train.drop('id', axis=1)
df_train
```

```python
x_train =df_train.values[:, 0:2]
y_train = df_train.values[:, 2]
```

```python
knn = KNeighborsRegressor(n_neighbors=3,metric='manhattan')
knn.fit(x_train, y_train)
```

```python
knn.predict([[5.5, 38]])
```

```python
l =knn.kneighbors([[5.5, 38]], n_neighbors=3, return_distance=False) # returns the nears k points not the distance
x_train[l].ravel() # point values
```
