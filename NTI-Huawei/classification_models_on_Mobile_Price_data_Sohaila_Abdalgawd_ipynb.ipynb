# **Support Vector Machine (SVM)**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/mobile_price_train.csv')
```

```python
df.info()
```

```python
df.shape
```

```python
corr_matrix = df.corr()
faeture_corr = corr_matrix[['price_range']].drop(index='price_range')
faeture_corr.sort_values(by='price_range', ascending=False)
```

```python
x= df[['battery_power', 'ram', 'px_width', 'px_height']]
y= df['price_range']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn import svm
model = svm.SVC(kernel = 'linear')
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression()
model_log.fit(x_train, y_train)
```

# **Decision Tree**


```python
from sklearn import tree
model_tree = tree.DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
```

```python
y_predict_tree = model_tree.predict(x_test)
accuracy_score(y_test, y_predict_tree)
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
model_forest = RandomForestClassifier()
model_forest.fit(x_train, y_train)
```

```python
y_predict_forest = model_forest.predict(x_test)
accuracy_score(y_test, y_predict_forest)
```

# **XGB**

```python
from xgboost import XGBClassifier
model_xgb = XGBClassifier()
model_xgb.fit(x_train, y_train)
```

```python
XGB_predict = model_xgb.predict(x_test)
accuracy_score(y_test, XGB_predict)
```
